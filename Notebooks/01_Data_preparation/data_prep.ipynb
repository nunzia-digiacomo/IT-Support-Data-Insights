```python
# Step 1: Importing Data

import pandas as pd

# Loading the Excel file
file_path = "IT_HelpDesk_Dataset.xlsx"
df_helpdesk = pd.read_excel(file_path)

# Basic info
print("Data loaded successfully!")
print("Shape:", df_helpdesk.shape)
print("Columns:", df_helpdesk.columns.tolist())
display(df_helpdesk.head())
```

```python
# Step 2: Analysing Data

# Quick overview
print("Data Types:")
print(df_helpdesk.dtypes)

print("\nBasic Info:")
df_helpdesk.info()

# Check first few rows
display(df_helpdesk.head())
```

```python
# Step 3: Checking for Missing Values, Duplicates, Anomalies, and Inconsistencies

# Missing values
missing = df_helpdesk.isnull().sum()
print("Missing Values per Column:")
print(missing[missing > 0])

# Duplicates
duplicates = df_helpdesk.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicates)

```

```python
# Basic anomaly checks: For numeric columns, check for extreme values
numeric_cols = df_helpdesk.select_dtypes(include='number').columns
for col in numeric_cols:
    print(f"\nColumn: {col}")
    print(df_helpdesk[col].describe())
```

```python
# Create a flag for unresolved tickets
df_helpdesk['In_Progress'] = df_helpdesk['Resolved Time'].isnull()

print(df_helpdesk[['Resolved Time', 'In_Progress']].head(10))
```

```python
# Check missing 'Item' by Category
missing_item_pattern = df_helpdesk[df_helpdesk['Item'].isnull()].groupby('Category')['Item'].count()
print("Missing Item counts by Category:")
print(missing_item_pattern)

# Check unresolved tickets by Category or Subcategory
in_progress_pattern = df_helpdesk[df_helpdesk['In_Progress']].groupby('Category')['Resolved Time'].count()
print("\nIn-progress tickets by Category:")
print(in_progress_pattern)
```

```python
# Inspect the missing 'Created Time' rows
print(df_helpdesk[df_helpdesk['Created Time'].isnull()][['Request ID', 'Subject', 'Created Time']])

df_helpdesk = df_helpdesk.dropna(subset=['Created Time'])
print("Rows after dropping missing Created Time:", df_helpdesk.shape[0])
```

```python
# Count unresolved tickets
in_progress_count = df_helpdesk['Resolved Time'].isnull().sum()
print("Total in-progress tickets:", in_progress_count)
```

```python
df_helpdesk['In_Progress'] = df_helpdesk['Resolved Time'].isnull()
```

```python
# Quick summary of ticket status
status_counts = df_helpdesk['In_Progress'].value_counts()
print("Ticket Status Counts (False = Closed, True = In Progress):")
print(status_counts)

# Open vs Closed tickets by Category
category_status = df_helpdesk.groupby('Category')['In_Progress'].sum()
print("Number of Open Tickets per Category:")
print(category_status)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot of open tickets per category
plt.figure(figsize=(12,6))
sns.barplot(x=category_status.values, y=category_status.index, palette='viridis')
plt.title("Number of Open Tickets per Category")
plt.xlabel("Open Tickets")
plt.ylabel("Category")
plt.show()
```
