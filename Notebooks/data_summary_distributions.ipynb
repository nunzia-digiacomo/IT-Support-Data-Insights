```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()
```

```python
# Basic descriptive stats
desc_stats = df[numeric_cols].describe().T
desc_stats['median'] = df[numeric_cols].median()
desc_stats['mode'] = df[numeric_cols].mode().iloc[0]
desc_stats['sum'] = df[numeric_cols].sum()
print("Descriptive statistics:")
display(desc_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
top_n = 10
max_len = 40  # truncate labels to first 40 characters

text_cols = ['Combined_Text', 'Tokenized_Text', 'Description']

for col in text_cols:
    plt.figure(figsize=(10,6))
    top_categories = df[col].value_counts().nlargest(top_n)
    labels = [str(l)[:max_len] + ('...' if len(str(l))>max_len else '') for l in top_categories.index]
    sns.barplot(x=top_categories.values, y=labels, palette='viridis')
    plt.title(f'Top {top_n} {col} Counts')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
for col in text_cols:
    print(f"\nTop {top_n} {col}:")
    display(df[col].value_counts().nlargest(top_n))
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```
